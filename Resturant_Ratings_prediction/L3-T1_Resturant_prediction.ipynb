```python
# city grouping 
city_proportion = df['City'].value_counts(normalize=True)
threshold = 0.01  # i.e., 1%
common_cities = city_proportion[city_proportion > threshold].index
df['City_grouped'] = df['City'].apply(lambda x: x if x in common_cities else 'Other')
```

```python
from sklearn.preprocessing import PowerTransformer

# Select the features
features = ['Longitude', 'Latitude']

# Apply PowerTransformer (Yeo-Johnson method handles negative values)
pt = PowerTransformer(method='yeo-johnson')

# Fit and transform the features
df[features] = pt.fit_transform(df[features])
```

```python
features = ['Average Cost for two', 'Votes']
for feature in features:
    # Plot boxplot
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()

    # Calculate IQR bounds
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    capped_feature_name = f"{feature}_capped"
    df[capped_feature_name] = df[feature].clip(lower=lower_bound, upper=upper_bound)

    # Optional: Print skewness after capping
    skewness = df[capped_feature_name].skew()
    print(f"Skewness of {capped_feature_name}: {skewness:.4f}")
```

<h2>Data Preprocessing</h2>

```python
df.info()
```

```python
# List of features to drop
features_to_drop = [
    'Restaurant ID', 'Restaurant Name','City','Locality', 'Address','Is delivering now', 'Rating color','Currency','Locality Verbose','City_grouped', 'Average Cost for two', 'Votes','Country Code','Cuisines','Switch to order menu'
]

# Drop the features
df.drop(columns=features_to_drop, inplace=True)
```

```python
df.info()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols] = df[numerical_cols].apply(pd.to_numeric, errors='coerce')
df[numerical_cols] = df[numerical_cols].replace([np.inf, -np.inf], np.nan)

# Impute if necessary
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_num = pd.DataFrame(
    scaler.fit_transform(df[numerical_cols]),
    columns=numerical_cols,
    index=df.index
)
```

```python
print(scaled_num.isnull().sum())
```

```python
from sklearn.preprocessing import StandardScaler

# Maping binary features to 0/1
binary_features = ['Has Table booking', 'Has Online delivery']
for col in binary_features:
    df[col] = df[col].map({'No': 0, 'Yes': 1})
    df[col] = pd.to_numeric(df[col], errors='coerce')
# Rating text to ordinal values
rating_text_mapping = {
    'Average': 2,
    'Not rated': -1,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5,
    'Poor': 1
}
df['Rating text'] = df['Rating text'].map(rating_text_mapping)

# Combing with binary and ordinal features
X_final = pd.concat([scaled_num, df[binary_features + ['Rating text']]], axis=1)
X_final = X_final.drop(columns=['Aggregate rating'], errors='ignore')

# Target variable
y = df['Aggregate rating']
```

<h2> Predictive Modeling</h2>

Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

model = RandomForestRegressor(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R-squared: {r2}")
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Evaluate
y_pred_lr = lr.predict(X_test)

mse_l = mean_squared_error(y_test, y_pred_lr)
rmse_l = np.sqrt(mse_l)
mae_l = mean_absolute_error(y_test, y_pred_lr)
r2_l = r2_score(y_test, y_pred_lr)

print(f"MSE: {mse_l}")
print(f"RMSE: {rmse_l}")
print(f"MAE: {mae_l}")
print(f"R-squared: {r2_l}")
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1]
}

xgb = XGBRegressor(objective='reg:squarederror', random_state=42)

grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,  
    scoring='neg_mean_squared_error',  
    verbose=1,
    n_jobs=-1  
)

grid_search.fit(X_final, y)
y_pred_x = grid_search.predict(X_test)
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

mse = mean_squared_error(y_test, y_pred_x)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_x)
r2 = r2_score(y_test, y_pred_x)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R-squared: {r2}")
```
